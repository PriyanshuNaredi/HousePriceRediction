```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
```

```python
URL = 'https://web.archive.org/web/20191218101945/https://housing.com/price-trends/property-rates-for-buy-in-indore_madhya_pradesh-P3fo3o3llgtfbceum'
response = requests.get(URL)
website_html = response.text
soup = BeautifulSoup(website_html, "html.parser")
```

```python
loc = []
avg_price=[]
for i in soup.find_all('span',class_='css-1y6emt2'):
    if i.text != 'Locality':

        loc.append(i.text)

for i in soup.find_all('span',class_='css-q3ye0i'):
    if i.text == 'avg_price' : 
        print()
    elif i.text == 'Avg. Price / Sqft':
        print()
    else:
        i = i.text
        i = i.replace("â‚¹",'')
        i = i.replace(",",'')
        i = float(i)
        avg_price.append(i)
```

```python
data = {'location':loc,'year':'2019','price_per_sqft':avg_price}
df1 = pd.DataFrame(data)
```

```python
df1.head()
```

```python
df1.insert(2,'2020',avg_price,True)
```
