```python
URL_2019 = 'https://web.archive.org/web/20191218101945/https://housing.com/price-trends/property-rates-for-buy-in-indore_madhya_pradesh-P3fo3o3llgtfbceum'
URL_2020 = 'https://web.archive.org/web/20200226141305/https://housing.com/price-trends/property-rates-for-buy-in-indore_madhya_pradesh-P3fo3o3llgtfbceum'
URL_2021 = 'https://web.archive.org/web/20210220221927/https://housing.com/price-trends/property-rates-for-buy-in-indore_madhya_pradesh-P3fo3o3llgtfbceum'
URL_2022 = ''
```

```python
response = requests.get(URL_2019)
website_html = response.text
soup = BeautifulSoup(website_html, "html.parser")
```

```python
loc = []
avg_price=[]
for i in soup.find_all('span',class_='css-1y6emt2'):
    if i.text != 'Locality':
        if i.text == 'L I G Colony':
            t = 'LIG Colony'
            loc.append(t)
        else:
            loc.append(i.text)

for i in soup.find_all('span',class_='css-q3ye0i'):
    if i.text == 'avg_price' : 
        print()
    elif i.text == 'Avg. Price / Sqft':
        print()
    else:
        i = i.text
        i = i.replace("₹",'')
        i = i.replace(",",'')
        i = float(i)
        avg_price.append(i)
```

```python
data = {'location':loc,'year':'2019','price_per_sqft':avg_price}
df1 = pd.DataFrame(data)
df1.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X = df[['location', 'year']]
y = df['price_per_sqft']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2
                                                    )
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown='ignore',sparse_output=False),['location'])
                                       ,remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
X_test
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
y_pred_lr
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lr
```

```python

ridge = Ridge()
pipe = make_pipeline(column_trans,scaler,ridge)
pipe.fit(X_train,y_train)
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```
